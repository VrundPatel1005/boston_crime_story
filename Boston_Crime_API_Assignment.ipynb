```
import requests
import urllib.parse
```

```
url = "https://data.boston.gov/api/3/action/datastore_search_sql"
```

```
sql = '''
SELECT
  "DISTRICT",
  "OCCURRED_ON_DATE",
  "SHOOTING",
  "OFFENSE_DESCRIPTION",
  "HOUR"
FROM "b973d8cb-eeb2-4e7e-99da-c92938efc9c0"
WHERE CAST("SHOOTING" AS INT) = 1
'''
```

```
# Build URL
final_url = url + "?sql=" + urllib.parse.quote(sql)


# Request data
rows = requests.get(final_url).json()

# Extract records
shootings = rows['result']['records']
```

```
print("Total records:", len(shootings))
print("First record:", shootings[0])
```

```
for s in shootings[:10]:
    print(s["DISTRICT"])
```

```
import pandas as pd
```

```
# Convert to DataFrame
df = pd.DataFrame(shootings)
```

```
# Clean data
df = df[df['DISTRICT'].notna()]
df['HOUR'] = df['HOUR'].astype(int)
```

```
#  Chart 1 data
district_counts = df.groupby('DISTRICT').size().reset_index(name='count')
```

```
# Chart 2 data
hour_counts = df.groupby('HOUR').size().reset_index(name='count')
```

```
# Show results
display(district_counts)
display(hour_counts)
```

```
# Save CSV files
district_counts.to_csv("district.csv", index=False)
hour_counts.to_csv("hour.csv", index=False)
```

```
# Bonus dataset for linking charts
district_hour = df.groupby(['DISTRICT','HOUR']).size().reset_index(name='count')
district_hour.to_csv("district_hour.csv", index=False)
```
